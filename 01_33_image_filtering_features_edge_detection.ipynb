Write some code to compute directional derivatives by steering the horizontal and vertical derivatives

$
f_{\theta} = cos(\theta)f_x + sin(\theta)f_y
$

```python
# 2.5: imagefiltering edges
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.signal import sepfir2d
from IPython.display import clear_output
from time import sleep

# compute space time derivatives
img = plt.imread("/content/alt.png")
p = [0.030320, 0.249724, 0.439911, 0.249724, 0.030320]
d = [-0.104550, -0.292315, 0.0, 0.292315, 0.104550]

if img.ndim == 3:  # If the image has 3 channels (e.g., RGB)
    img = np.mean(img, axis=2)  # Convert to grayscale by averaging channels

img_x = sepfir2d( img, d, p ) # spatial (x) derivative
img_y = sepfir2d( img, p, d ) # spatial (y) derivative

for th in range(0,361,5):
  plt.figure(figsize=(10,10))
  plt.imshow( np.cos(np.radians(th))*img_x + np.sin(np.radians(th))*img_y, cmap='gray' )
  plt.title(th)
  plt.show()
  sleep(0.1)
  clear_output(wait=True)
```
