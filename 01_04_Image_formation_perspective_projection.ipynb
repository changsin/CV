### 1.6.2 2D rigid body transformation

In a 2D plane, a point with coordinates (x,y) can be transformed to a new point (xâ€² ,yâ€²) using the following formulas:

$$ x' = x \ cos\theta - y \ sin\theta + t_x $$

$$ y' = x \ sin\theta + y \ cos\theta + t_y $$

- $ \theta $: the angle of rotation
- $ t_x, t_y$: translation amounts in the x and y directions

### 1.6.7 Affine Transformation
An affine transformation is a type of geometric transformation that preserves parallelism but not necessarily distances or angles. This makes it more general than a Euclidean (rigid body) or similarity transformation.

It combines the following operations:
- Rotation: Changing the orientation.
- Non-uniform Scaling: Stretching or compressing an object differently along the x and y axes.
- Shearing: Skewing an object.
- Translation: Shifting the position.

$$
\begin{bmatrix}
x' \\
y' \\
1
\end{bmatrix}
=
\begin{bmatrix}
a_{11} & a_{12} & t_x \\
a_{21} & a_{22} & t_y \\
0 & 0 & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y \\
1
\end{bmatrix}
$$

The key property is that if two lines are parallel before the transformation, they will remain parallel after the transformation, but their spacing and the angles between lines may change.

The 2Ã—2 matrix
$
 \begin{bmatrix}
a_{11} & a_{12} \\
a_{21} & a_{22}
\end{bmatrix}
$
 is the linear transformation matrix. This block combines the effects of rotation, scaling, and shearing.

The first column
$
\begin{bmatrix}
a_{11} \\
a_{21}
\end{bmatrix}
$
 describes the new, transformed x-axis.

The second column$
\begin{bmatrix}
a_{12} \\
a_{22}
\end{bmatrix}
$
 describes the new, transformed y-axis.

 The last column
$
\begin{bmatrix}
t_x \\
t_y
\end{bmatrix}
$ is the translation vector, which shifts the object.

### 1.6.7 Projective Transformation (homography)

A projective transformation, also known as a homography, is the most general type of 2D linear geometric transformation. It is the mathematical model for how a flat plane in 3D space is transformed when viewed by a camera.

Unlike affine transformations, a projective transformation does not preserve parallelism. It's what causes parallel lines in the real world (like railroad tracks) to appear to converge at a vanishing point in a photograph. The only property it preserves is the straightness of lines; a straight line in the original plane remains a straight line after the transformation.

$$
\begin{bmatrix}
x' \\
y' \\
1
\end{bmatrix}
=
\begin{bmatrix}
h_{11} & h_{12} & h_{13} \\
h_{21} & h_{22} & h_{23} \\
h_{31} & h_{32} & h_{33}
\end{bmatrix}
\begin{bmatrix}
x \\
y \\
1
\end{bmatrix}
$$


The key difference from affine transformations is the third row of the matrix, specifically the non-zero values for $ h_{31} $ and $ h_{32} $.

When you perform the matrix multiplication, you get:

- $ x' = h_{11} + h_{12} + h_{13} $

- $ y' = h_{21} + h_{22} + h_{23} $

- $ w' = h_{31} + h_{32} + h_{33} $

To get the final Cartesian coordinates, you must divide the first two components by the third one ($ w' $):

$$
x'' = \frac{x'}{w'} = \frac{h_{11}x + h_{12}y + h_{13}}{h_{31}x + h_{32}y + h_{33}}
$$

$$
y'' = \frac{y'}{w'} = \frac{h_{21}x + h_{22}y + h_{23}}{h_{31}x + h_{32}y + h_{33}}
$$

This division is what creates the perspective effect. The value of $ w' $ changes depending on the original point $ (x,y) $, causing some points to be scaled differently than others, which is the essence of perspective.

##01-04 [Image Formation: pinhole camera: perspective projection, 2-D](https://youtu.be/iMrRwXtab6Y?si=SrGbpFi3v2a35dYp&t=480).

Write some code that simulates the projection of a line segment defined by two points (-5, Z) and (5, Z), where Z ranges from 10 to 1,000, and assuming a camera focal length of f=1.

For each distance Z, project the two points into a 1-D sensor under perspective projection, and compute the length of the segment.

Plot this length as a function of distance Z to see how size changes as a function of distance to the camera.

```python
# prompt: Calculate the perspective projection length of two points pt1 and pt2 as a function of Z: Use the two points pt1=(-5, Z) pt2=(5, Z). The formula is x = - X/Z

def get_perspective_projection_length(pt1, pt2):
  """
  Calculate the perspective projection length of two points pt1 and pt2 as a function of Z.
  The formula is x = - X/Z.

  Args:
    pt1: A tuple representing the first point (x, z).
    pt2: A tuple representing the second point (x, z).

  Returns:
    The perspective projection length.
  """
  x1, z1 = pt1
  x2, z2 = pt2

  x1_projected = -x1 / z1
  x2_projected = -x2 / z2

  return abs(x2_projected - x1_projected)


# Example usage
pt1 = (-5, 10)  # x, z
pt2 = (5, 10)  # x, z

length = get_perspective_projection_length(pt1, pt2)
print(f"Perspective projection length: {length}")
```

```python
# prompt: Can you plot it as well?

import matplotlib.pyplot as plt
import numpy as np

def get_perspective_projection_length(pt1, pt2):
  """
  Calculate the perspective projection length of two points pt1 and pt2 as a function of Z.
  The formula is x = - X/Z.

  Args:
    pt1: A tuple representing the first point (x, z).
    pt2: A tuple representing the second point (x, z).

  Returns:
    The perspective projection length.
  """
  x1, z1 = pt1
  x2, z2 = pt2

  x1_projected = -x1 / z1
  x2_projected = -x2 / z2

  return abs(x2_projected - x1_projected)


# Example usage
pt1 = (-5, 10)  # x, z
pt2 = (5, 10)  # x, z

length = get_perspective_projection_length(pt1, pt2)
print(f"Perspective projection length: {length}")

# Create a range of Z values
z_values = np.linspace(10, 1000, 1000)

# Calculate the projection length for each Z value
lengths = []
for z in z_values:
  pt1_z = (-5, z)
  pt2_z = (5, z)
  lengths.append(get_perspective_projection_length(pt1_z, pt2_z))

# Plot the results
plt.plot(z_values, lengths)
plt.xlabel("Z")
plt.ylabel("Perspective Projection Length")
plt.title("Perspective Projection Length vs. Z")
plt.grid(True)
plt.show()
```
